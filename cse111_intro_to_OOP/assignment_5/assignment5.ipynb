```python
#task_1
import math 
class Circle():
  def __init__(self,rad):
    self.__rad = rad
  def area(self):
    area = math.pi * self.__rad ** 2
    return f"{area}"
  def getRadius(self):
    return f"{self.__rad}"

c1 = Circle(4)
print("First circle radius:" , c1.getRadius())
print("First circle area:" , c1.area())


c2 = Circle(5)
print("Second circle radius:" , c2.getRadius())
print("Second circle area:" , c2.area())


```

```python
#task_2

class Triangle():
    def __init__(self,base,height):
        self.__base = int(base)
        self.__height = int(height)
    def getBase(self):
        return f"{self.__base}"
    def getHeight(self):
        return f"{self.__height}"
    def area(self):
        return f"{.5 * self.__height * self.__base}"

t1 = Triangle(10, 5)
print("First Triangle Base:" , t1.getBase())
print("First Triangle Height:" , t1.getHeight())
print("First Triangle area:" ,t1.area())
t2 = Triangle(5, 3)
print("Second Triangle Base:" , t2.getBase())
print("Second Triangle Height:" , t2.getHeight())
print("Second Triangle area:" ,t2.area())
    
```

```python
#task 6

class Train:
  def __init__(self, name, *stations):
    self.name = name
    self.stations = list(stations)
    self.passengers = []
    print(f'Welcome aboard on {name}')
    print(f'Start: {stations[0]}')
    print(f'Destination: {stations[-1]}')

  def addPassenger(self, *passengers):
    passengers = list(passengers)
    for passenger in passengers:
      self.passengers += [passenger]
      print(f'{passenger.p_name} welcome aboard')

  def allPassengerDetails(self):
    for i in self.passengers:
      print(f'Name: {i.p_name}', end=' ')
      if len(i.stops) == 0:
        i.start = self.stations[0]
        i.stop = self.stations[-1]
      elif len(i.stops) == 1:
        i.start = i.stops[0]
        i.stop = self.stations[-1]
      elif len(i.stops) == 2:
        i.start = i.stops[0]
        i.stop = i.stops[1]
      print(f'Start: {i.start}', end = ' ')
      print(f'Stop: {i.stop}', end = ' ')
      i.fare = (self.stations.index(i.stop) - self.stations.index(i.start)) * 100
      print(f'Fair: ${i.fare}')

class Passenger:
  def __init__(self, p_name, *stops):
    self.p_name = p_name
    self.stops = list(stops)

t1 = Train('T1-Express','New York','Manhattan','Brooklyn','Boston')
print("1========================")
p1 =Passenger("Naruto")
t1.addPassenger(p1)
p2 = Passenger("Sasuke","Manhattan")
p3 = Passenger("Hinata","Manhattan","Brooklyn")
print("2========================")
t1.addPassenger(p2,p3)
print("3========================")
t1.allPassengerDetails()
print("4========================")
t2 = Train('Europe-Express','London','Paris','Brussels','Turkey')
print("5========================")
p4 =Passenger("Max","London","Brussels")
p5 = Passenger("Eleven","Paris")
p6 = Passenger("Mike","Brussels")
t2.addPassenger(p4,p5,p6)
print("6========================")
t2.allPassengerDetails()          
       
           
         
```

```python
#task 7

class BracuStudent:
  def __init__(self, name, home):
    self.name = name
    self.home = home
    self.b_pass = False
  
  def get_pass(self):
    self.b_pass = True

  def show_details(self):
    print(f'Student Name: {self.name}')
    print(f'Lives in {self.home}')
    print(f'Have Bus Pass? {self.b_pass}')

class BracuBus:
  def __init__(self, route, max = 2):
    self.route = route
    self.max = max
    self.passenger = []

  def board(self, *students):
    if len(students) > 0:
      for st in students:
        if st.b_pass == True:
          if st.home == self.route:
            self.passenger += [st.name]
            print(f'{st.name} boarded the bus')
            if len(self.passenger) == self.max:
              print(f'Bus is full!')
              break
          else:
            print(f'You got on wrong bus!')
        else:
          print(f"You don't have bus pass!")
    elif len(students) == 0:
      print(f'No passenger!')

  def show_details(self):
    print(f'Bus Route: {self.route}')
    print(f'Passengers Count: {len(self.passenger)} (Max: {self.max})')
    print(f'Passengers On Board: {self.passenger}')

st1 = BracuStudent("Afif", "Mirpur")
print("1===========================")
st2 = BracuStudent("Shanto", "Motijheel")
st3 = BracuStudent("Taskin", "Mirpur")
st1.show_details()
st2.show_details()
print("2===========================")
st3.show_details()
print("3===========================")
bus1 = BracuBus("Mirpur")
bus2 = BracuBus("Azimpur", 5)
bus1.show_details()
bus2.show_details()
print("4===========================")
st2.get_pass()
st3.get_pass()
print("5===========================")
st2.show_details()
st3.show_details()
print("6===========================")
bus1.board()
print("7===========================")
bus1.board(st1, st2)
print("8===========================")
st1.get_pass()
st2.home = "Mirpur"
st1.show_details()
st2.show_details()
print("9===========================")
bus1.board(st1, st2, st3)
print("10===========================")
bus1.show_details()
```

```python
#task 8

class Library:
  def __init__(self, name, book_dict):
    self.name = name
    self.book_dict = book_dict
    self.borrower = {}

  def details(self):
    print(f'{self.name} Library details')
    print(f'Borrower details:')
    print(f'{self.borrower}')
    print(f'Books availability:')
    print(f'{self.book_dict}')

class Reader:
  def __init__(self, r_name):
    self.r_name = r_name
    self.book_list = {}
    self.total = 0

  def borrow(self, obj, *genre):
    self.obj = obj
    self.genre = genre
    for book in genre:
      if self.total < 5:
        if obj.book_dict[book] > 0:
          print(f'{book} book is borrowed successfully.')
          if book not in self.book_list:
            self.book_list[book] = 1
            self.total += 1
          else:
            self.book_list[book] += 1
            self.total += 1
        elif obj.book_dict[book] == 0:
          print(f'{book} books are not available at the moment.')
      elif self.total == 5:
        print(f'You cannot borrow more than 5 books.')
    obj.borrower.update({self.r_name: self.total})
    for key, val in obj.book_dict.items():
      if key in genre:
        if val > 0:
          val -= 1
          obj.book_dict.update({key: val})
   
  def readerInfo(self, *type):
    if len(type) == 0:
      print(f'{self.r_name}, you have {len(self.book_list)} book(s) with you.')
      for key, val in self.book_list.items():
        print(f'Books on {key}: {val}')
    elif len(type) == 1:
      print(f'{self.r_name}, you have {self.book_list[type[0]]} {type[0]} book(s) with you') 

L1=Library('Dhaka',{'Arts':15,'Fiction':135,'Politics':2,'Science':11,'Poetry':15})
L1.details()
print("1----------------------")
r1=Reader('Aladdin')
r1.borrow(L1,'Arts','Fiction','Fiction','Politics')
print("2----------------------")
r1.borrow(L1,'Politics','Fiction')
print("3----------------------")
r1.readerInfo()
print("4----------------------")
r1.readerInfo('Fiction')
print("5----------------------")
L1.details()
print("6----------------------")
r2=Reader('Jasmine')
r2.borrow(L1,'Politics','Poetry')
print("7----------------------")
r2.readerInfo()
print("8----------------------")
L1.details()
```
