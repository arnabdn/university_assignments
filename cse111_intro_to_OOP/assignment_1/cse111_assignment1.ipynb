```python
#st_task2

inp = input("Enter string")
num_count = 0
word_count = 0

for i in inp:
  if 97<=ord(i)<122 or 65<=ord(i)<=97:
    word_count += 1
  if 48<=ord(i)<=57:
    num_count += 1

if word_count and num_count:
  print("MIXED")
elif word_count >= 1 and num_count == 0:
  print("WORD")
elif num_count >= 1 and word_count == 0:
  print("NUMBER")
```

```python
#st_task5 c

txt = input("put your password here")
txt2 = ""
up = 0
low = 0
num = 0 
sp = 0

for i in txt:
  if 65<=ord(i)<=90:
    up += 1
  elif 90<=ord(i)<=122:
    low += 1
  elif 48<=ord(i)<=57:
    num += 1
  else:
    sp += 1

if (up and low and num and sp) >= 1:
  print("OK")
if up == 0:
  txt2 += "Uppercase Missing, "
if low == 0:
  txt2 += "Lowercase Missing, "
if num == 0:
  txt2 += "Digit Missing, "
if sp == 0:
  txt2 += "Special Missing, "

print(txt2[0:-2])
```

```python
#lst_task2

n = int(input())
lst1 = []
lst2 = []
high = -99999999
sum = 0
idx = 0
for i in range(n):
    txt = input().split()
    for j in txt:
        lst2 += [int(j)]
    lst1 += [lst2]
    lst2 = []

for i in range(len(lst1)):
    for j in lst1[i]:
        sum += j
    if sum > high:
        high = sum
        idx = i
    sum = 0
print(high)
print(lst1[idx])
```

```python
#lst_task4

lst = []
lst2 = []
n = 0

while(True):
  inp = input()
  if inp == 'STOP':
    break
  for i in inp.split():
    lst += [int(i)]
  for j in range(1,len(lst)):
    diff = abs(lst[j] - lst[j - 1])
    lst2 += [diff]
  for k in range(1,len(lst)):
    if k in lst2:
      n += 1
  if n == (len(lst) - 1):
    print("UB Jumper")
  else:
    print("Not UB Jumper")
  lst = []
  lst2 = []
  n = 0

```

```python
#lst_task5

nk = input("enter n k").split()
n = int(nk[0])
k = int(nk[1])
int_lst = []
tmp_lst = []
final_lst = []
added_lst = []
team = 0

lst = input().split()
for i in lst:
    int_lst += [int(i)]

srt_lst = sorted(int_lst)
div = n//3

for i in range(len(srt_lst)):
    srt_lst[i] += k

for i in range(div):
    for j in range(len(srt_lst)):
        tmp_lst += [srt_lst[j]]
        if len(tmp_lst) == 3:
            final_lst += [tmp_lst]
            tmp_lst = []
        if len(final_lst) == div:
            break

for i in range(len(srt_lst)):
    if srt_lst[i] > 5:
        continue
    else:
        team += 1

print(team//3)
```

```python
#dic_task1

dic_a = {"a": 100, "b": 100, "c": 200, "d": 300}
dic_b = {"a": 300, "b": 200, "d": 400, "e": 200}
merged_dic = {**dic_a,**dic_b}

for i,j in dic_a.items():
    if i in dic_b:
        merged_dic[i] += j

lst = list(set(merged_dic.values()))
print(merged_dic)
print(f"Values: {tuple(sorted(lst))}")
```

```python
#dic_task2

times = {}

while(True):
    num = input("enter your number: ")
    if num == "STOP":
        break
    if num not in times:
        times[num] = 1
    elif num in times:
        times[num] += 1

for i,j in times.items():
    print(f"{i} - {j} times")
```
