```python
#task1 

class Customer():
    def __init__(self,name):
        self.name = name
    
    def greet(self,name = None):
        if name == None:
            print(f"Hello!")
        else:
            print(f"Hello {name}!")
    def purchase(self,*foods):
        print(f"{self.name} you purchased {len(foods)} item(s)")
        for food in foods:
            print(food)
        

customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
#task 2

class Panda():
    def __init__(self,name,gender,age):
        self.name = name
        self.gender = gender
        self.age = age
    def sleep(self,sleep = None):
        if sleep == None:
            return f"{self.name}s duration is unknown and thus should have only bamboo leaves"
        if 3<=sleep<=5:
            return f"{self.name} sleeps {sleep} hours daily and should have Mixed Veggies"
        elif 6<=sleep<=8:
            return f"{self.name} sleeps {sleep} hours daily and and should have Eggplant and Tofu"
        elif 9<=sleep<=11:
            return f"{self.name} sleeps {sleep} hours and should have Broccooli and Chicken"

panda1 = Panda("Kunfu", "Male", 5)
panda2 = Panda("Pan Pan", "Female",3)
panda3 = Panda("Ming Ming", "Female",8)
print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age))
print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age))
print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))
print("===========================")
print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())
```

```python
#task 4

class Student():
    def __init__(self,name = "default student"):
        self.name = name
        self.score = 0
    def quizcalc(self,*numbers):
        total = 0
        for number in numbers:
            total += number
        self.score = total
        
    def printdetail(self):
        print(f"Hello {self.name}")
        print(f"Your average quiz score is {self.score/3}")

s1 = Student()
s1.quizcalc(10)
print('--------------------------------')
s1.printdetail()
s2 = Student('Harry')
s2.quizcalc(10,8)
print('--------------------------------')
s2.printdetail()
s3 = Student('Hermione')
s3.quizcalc(10,9,10)
print('--------------------------------')
s3.printdetail()
 
 
```

```python
#task 6

class Patient():
    def __init__(self,name,age):
        self.name = name
        self.age = age
        self.symptoms = []
    def add_Symptom(self,*symptoms):
        for symptom in symptoms:
            self.symptoms += [symptom]
    def printPatientDetail(self):
        print(f"Name: {self.name}")
        print(f"age: {self.age}")
        symptoms = ""
        for i in self.symptoms:
            symptoms += f"{i}, "
        print(f"Symptoms: {symptoms[:-2]}")

p1 = Patient('Thomas', 23)
p1.add_Symptom('Headache')
p2 = Patient('Carol', 20)
p2.add_Symptom('Vomiting', 'Coughing')
p3 = Patient('Mike', 25)
p3.add_Symptom('Fever', 'Headache', 'Coughing')
print("=========================")
p1.printPatientDetail()
print("=========================")
p2.printPatientDetail()
print("=========================")
p3.printPatientDetail()
```

```python
#task 7

class Match():
    def __init__(self,vs):
        self.vs = vs.split("-")
        self.batting = self.vs[0]
        self.bowling = self.vs[1]
        self.runs = 0
        self.wickets = 0
        self.overs = 0
        print("5..4..3..2..1.. Play !!!")
    def add_run(self,run):
        if 0<=run<=6:
            self.runs += run
    def add_wicket(self,wicket):
        self.wickets += wicket
    def add_over(self,over):
        if (over + self.overs) <= 5:
            self.overs += over
        else:
            print(f"Warning!cannot add {over} over/s (5 over match)")
    def print_scoreboard(self):
        print(f"Batting Team: {self.batting}")
        print(f"Bowling Team: {self.bowling}")
        return f"Total Runs: {self.runs} Wickets: {self.wickets} Overs: {self.overs}"

match1 = Match("Rangpur Riders-Cumilla Victorians")
print("=========================")
match1.add_run(4)
match1.add_run(6)
match1.add_over(1)
print(match1.print_scoreboard())
print("=========================")
match1.add_over(5)
print("=========================")
match1.add_wicket(1)
print(match1.print_scoreboard())

```

```python
#task 08

class ParcelKoro():
    def __init__(self,name = None,weight = 0):
        self.name = name
        self.product_weight = weight
        self.fee = 0
        self.location_charge = 50
    
    def calculateFee(self,location = None):
        if location != None:
            self.location_charge = 100
        if self.product_weight == 0:
            self.fee = 0
        else:
            self.fee = (self.product_weight)*20 + self.location_charge

        
    def printDetails(self):
        if self.name == None:
            print("Customer Name: No name set")
        else:
            print(f"Customer Name: {self.name}")
        print(f"Product weight: {self.product_weight}")
        print(f"Total fee:{self.fee}")

print("**********************")
p1 = ParcelKoro()
p1.calculateFee()
p1.printDetails()
print("**********************")
p2 = ParcelKoro('Bob The Builder')
p2.calculateFee()
p2.printDetails()
print("----------------------------")
p2.product_weight = 15
p2.calculateFee()
p2.printDetails()
print("**********************")
p3 = ParcelKoro('Dora The Explorer', 10)
p3.calculateFee('Dhanmondi')
p3.printDetails()
```

```python
#task 9

class Batsman():
    def __init__(self,name,runs,balls = None):
        if type(name) != str:
            self.name = "New Batsman"
            self.runs = name
            self.balls = runs
        else:
            self.name = name
            self.runs = runs
            self.balls = balls
    def printCareerStatistics(self):
        print(f"Name: {self.name}")
        print(f"Runs Scored: {self.runs}, Balls Faced: {self.balls}")
    def battingStrikeRate(self):
        return f"{self.runs/self.balls * 100}"
    def setName(self,name):
        self.name = name


b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
#task 12

class TaxiLagbe():
    def __init__(self,number,area):
        self.number = number
        self.area = area
        self.names = []
        self.fare = 0
        self.passenger_count = 0
                  
    def addPassenger(self,*passengers):
        if self.passenger_count == 4:
            print(f"Taxi Full! No more passengers can be added.")
        else:
            for passenger in passengers:
                name,fare = passenger.split("_")
                self.names += [name]
                self.fare += int(fare)
                self.passenger_count += 1
                print(f"Dear {name}! Welcome to Taxilagbe.")
    def printDetails(self):
        print(f"Trip info for Taxi number: {self.number}")
        print(f"This taxi can cover only in {self.area} area")
        print(f"Total passenger {self.passenger_count}")
        print(f"Passenger lists: ")
        passengers = ""
        for passenger in self.names:
            passengers += f"{passenger}, "
        print(f"{passengers[0:-2]}")
        print(f"Total collected fare: {self.fare}")


taxi1 = TaxiLagbe('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100',
'Wood_200')
taxi1.addPassenger('Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115')
taxi2.addPassenger('Parker_215')
print('-------------------------------')
taxi2.printDetails()


```

```python
#task 13

class Account:
  def __init__(self, name = 'Default account', balance = 0.0):
    self.name = name
    self.balance = balance
  def withdraw(self, amount):
    self.amount = amount
    if self.balance - amount <= 3070:
      print('Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.')
    elif self.balance - amount > 3070:
      print('Withdraw successful! New balance is:')
      print(self.balance - amount)
  def details(self):
    return self.name + "\n" + str(self.balance)
  
a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930)
print("------------------------")
a2.withdraw(600)
print("------------------------")
a1.withdraw(6929)
```

```python
#task 14

class StudentDatabase():
    def __init__(self,name,id):
        self.name = name
        self.id = id
        self.grades = {}

    def calculateGPA(self,lst,sem):
        courses = []
        attempted_creds = 0
        sum = 0
        for i in lst:
            course_grade = i.split(": ")
            courses += [course_grade[0]]
            sum += float(course_grade[1]) * 3
            attempted_creds += 3
        grade = str((sum / attempted_creds))
        d2 = {tuple(courses):grade}
        d1 = {sem:d2}
        for key,value in d1.items():
            self.grades[key] = value
    def printDetails(self):
        print(f"Name: {self.name}")
        for semester,data in self.grades.items():
            print(f"Courses taken in {semester}:")
            for courses,cg in data.items():
                for course in courses:
                    print(course)
            if len(cg) > 3:
                cg = cg[:4]
                if cg[-1] == "0":
                    print(f"GPA: {cg[0:3]}")
                else:
                    print(f"GPA: {cg}")
            else:
                print(f"GPA: {cg}")
                

s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 
'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021') 
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------') 
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 
'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------') 
s2.printDetails()
```
